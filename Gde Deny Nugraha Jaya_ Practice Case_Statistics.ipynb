#PREPROCESSING DATA

Choose data according to the question criteria, namely data in 2017

```python
datastat17=datastat.loc[datastat['Year']==2017]
```

View the variable type of each data

Check Missing Value

```python
datastat17.isnull().sum()
```

```python
#Delete column blan1 and blank2
datastat17=datastat17.drop(['blanl','blank2'], axis = 1)
```

```python
#Delete row missing
datastat17=datastat17.dropna()
```

Check data duplication by player name

```python
datastat17.duplicated('Player').sum()
```

```python
#drop duplicate player
datastat17=datastat17.drop_duplicates(subset=['Player'], keep='last')
```

1. Who is the youngest and oldest player in the NBA in 2017 for each team (Tm) ?

```python
datamax=datastat17.groupby('Tm')[['Player','Age']].max()
datamax
```

```python
datamin=datastat17.groupby('Tm')[['Player','Age']].min()
datamin
```

2.	Which player has the most minutes played (MP) in each position (Pos)? 

```python
mostminute=datastat17.groupby('Pos')[['Player','MP']].max()
mostminute
```

3.	Which team has the highest average total rebound percentage (TRB%), assist percentage (AST%), steal percentage (STL%), and block percentage (BLK%)? 

```python
#initialization max value from each variable
teammax=datastat17.groupby('Tm')[['TRB%','AST%','STL%','BLK%']].mean()
maxtrb=max(teammax['TRB%'])
maxast=max(teammax['AST%'])
maxstl=max(teammax['STL%'])
maxblk=max(teammax['BLK%'])
```

```python
teammax.loc[teammax['TRB%']==maxtrb]
```

```python
teammax.loc[teammax['AST%']==maxast]
```

```python
teammax.loc[teammax['STL%']==maxstl]
```

```python
teammax.loc[teammax['BLK%']==maxblk]
```
